```python
def merge_tudatasets_new(dataset1, dataset2):
    """
    
    """
    # 
    aug1, aug_ratio1 = dataset1.aug, dataset1.aug_ratio
    aug2, aug_ratio2 = dataset2.aug, dataset2.aug_ratio
    dataset1.aug, dataset1.aug_ratio = 'none', None
    dataset2.aug, dataset2.aug_ratio = 'none', None

    # 
    data_list = [
        dataset1.get(i) for i in range(len(dataset1))
    ] + [
        dataset2.get(i) for i in range(len(dataset2))
    ]

    # 
    merged = TUDatasetExt.__new__(TUDatasetExt)
    
    # 
    merged.root = ""  # 
    merged.name = f"merged_{dataset1.name}_{dataset2.name}"
    merged.use_node_attr = True
    merged.pre_transform = dataset1.pre_transform
    merged.processed_filename = "memory_data.pt"  # 
    merged.aug = "none"
    merged.aug_ratio = None
    merged.transform = None
    merged.pre_filter = None
    merged.__indices__ = None
    
    # 
    merged.data, merged.slices = merged.collate(data_list)
    
    # 
    for attr in ['p_edge_node', 'deg_max', 'deg_min']:
        if hasattr(dataset1, attr):
            setattr(merged, attr, getattr(dataset1, attr))
    
    # 
    dataset1.aug, dataset1.aug_ratio = aug1, aug_ratio1
    dataset2.aug, dataset2.aug_ratio = aug2, aug_ratio2
    
    return merged
```
